```python
# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Ignore useless warnings (see SciPy issue #5998)
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")

import pandas as pd

from sklearn.model_selection import train_test_split

data= pd.read_csv("data12.csv", dtype=float)
data=data.dropna(axis=0,how='any')

data_phr={}
dataset={}
train_set={}
test_set={}
train_x={}
train_y={}
test_x={}
test_y={}
data_x={}
data_y={}
data_join={}

predict_arr={}
final_predict_arr={}
```

```python
for i in range(6):
   
    
    data_phr[i]= pd.read_csv("phr.csv",usecols=[i+2])
    data_phr[i]=data_phr[i].dropna(axis=0,how='any')
    data_phr[i][data_phr[i]<0.2+i]=0
    data_phr[i][data_phr[i]>=0.2+i]=1

    data_join[i]= data.join(data_phr[i])
    data_join[i]= data_join[i].dropna(axis=0,how='any')

    dataset[i]= data_join[i].drop(['x','y','d1ndwi','d2ndwi','d1evi','phreat','d2evi','d1prec','d2prec','d1ndvi','d2ndvi','d1temp','d2temp','evapot','elevation','gwlvl','t1','t2','d1sat0','d2sat0','d1sat1','d2sat1','d1sat2','d2sat2','d1sat3','d2sat3','d1sat4','d2sat4'], axis=1)
    #bdataset[i]= dataset[i].join(data_phr[i])

    train_set[i], test_set[i] = train_test_split(dataset[i], test_size=0.2)
    print(len(train_set[i]), "train" + str(i) + " +", len(test_set[i]), "test"+ str(i))

    train_x[i], train_y[i] = train_set[i], train_set[i].pop('ph'+str(i+1))
    test_x[i], test_y[i] = test_set[i], test_set[i].pop('ph'+str(i+1))

    import tensorflow as tf
    feature_columns = []
    
    for key in train_x[i].keys():
        feature_columns.append(tf.feature_column.numeric_column(key=key))

    classifier = tf.estimator.DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[50,100,50],
                                            n_classes=2,
                                            optimizer=tf.train.AdamOptimizer(
                                            learning_rate=0.01)
                                           )
    def train_func(train_x,train_y):
        dataset=tf.data.Dataset.from_tensor_slices((dict(train_x), train_y))
        dataset = dataset.shuffle(100).repeat().batch(100)
        return dataset

    def eval_input_fn(features, labels, batch_size):
        features=dict(features)
        if labels is None:
            # No labels, use only features.
            inputs = features
        else:
            inputs = (features, labels)
        dataset = tf.data.Dataset.from_tensor_slices(inputs)

        assert batch_size is not None, "batch_size must not be None"
        dataset = dataset.batch(batch_size)
        return dataset
    
    classifier.train(input_fn=lambda:train_func(train_x[i],train_y[i]),steps=200)

    predict_arr[i] = []
    predictions = classifier.predict(input_fn=lambda:eval_input_fn(test_x[i],labels=test_y[i],batch_size=100))

    for predict in predictions:
        predict_arr[i].append(predict['probabilities'].argmax())

    result = predict_arr[i] == test_y[i]
    result1 = [w for w in result if w == True]
    print("Accuracy"+str(i+1)+" is %s"%str((len(result1)/len(result))))
    
    data_x[i], data_y[i] = dataset[i], dataset[i].pop('ph'+str(i+1))
    
    final_predict_arr[i] = []
    predictions = classifier.predict(input_fn=lambda:eval_input_fn(data_x[i],labels=data_y[i],batch_size=100))

    for predict in predictions:
        final_predict_arr[i].append(predict['probabilities'].argmax())

    result = final_predict_arr[i] == data_y[i]
    result1 = [w for w in result if w == True]
    print("Accuracy"+str(i+1)+" is %s"%str((len(result1)/len(result))))
    
    data.plot(kind="scatter",x="y",y="x",alpha=0.4,s=1,c=final_predict_arr[i])
    plt.legend()
    
    data.plot(kind="scatter",x="y",y="x",alpha=0.4,c=data_y[i])
    plt.legend()
```

```python
data.plot(kind="scatter",x="y",y="x",alpha=0.4,s=1,c=final_predict_arr[0])
plt.legend()
data.plot(kind="scatter",x="y",y="x",alpha=0.4,c=data_y[0])
plt.legend()
```
